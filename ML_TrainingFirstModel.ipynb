#Loading CSV dataset using pandas

```python
import pandas as pd
df = pd.read_csv("/content/iris.csv",on_bad_lines='skip')
df.head(5)
df.isna().sum()
df = df.dropna()
df.isna().sum()
df1 = df.replace({'setosa':'0','versicolor':'1','virginica':'2'})
df1['Species']=pd.to_numeric(df1['Species'])
#for ind in df.index:

 # if(df['Species'][ind]=='setosa'):
    #df['Species'][ind] = 0.0

df1.tail(2)
```

Above dataset has multiple inputs like length, width etc to identify species of the flower.

#Data preperation
Saperate dataset into Xs ie inputs and Y ie output

##Output dataframe

```python
y_df = df1[['Species']]
y_df.head(5)
```

##Input dataframe

```python
#axis = 1 means work on rows, 0 if want to work with columns
x_df = df1.drop('Species', axis=1)
x_df.head(2)
```

##Convert non numeric values to numeric

##Splitting dataset in training and test dataset
In Scikit-learn, train_test_split is a function used to create training and testing data to be used to measure a machine learning model’s performance.

training dataset size = 80% and test dataset size = 20%

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_df,y_df,test_size=0.2)
y_train.head(5)
```

# "Linear Ridge classifier" model training

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

#Apply model to make prediction

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

#Model Performance evaluation

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
lr_train_mse = mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2 = r2_score(y_train,y_lr_train_pred)
print(lr_train_mse)
print(lr_train_r2)

lr_test_mse = mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2 = r2_score(y_test,y_lr_test_pred)
print(lr_test_mse)
print(lr_test_r2)
```
